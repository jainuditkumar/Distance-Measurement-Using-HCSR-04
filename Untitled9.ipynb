```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.stats.diagnostic import het_breuschpagan
from scipy.stats import shapiro
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

## PART 1: Linear Regression on Housing Data

#### **Correlation Matrix**
We begin by visualizing the correlation matrix between numerical features in the housing dataset. This helps us identify relationships between variables. For instance, we observe a strong positive correlation between `median_income` and `median_house_value`, indicating that wealthier areas tend to have higher housing values. Other features like `total_rooms`, `total_bedrooms`, and `households` also show multicollinearity.

```python
housing_df = pd.read_csv('housing.csv')
housing_df_clean = housing_df.dropna().copy()
housing_df_clean = pd.get_dummies(housing_df_clean, columns=['ocean_proximity'], drop_first=True)
housing_df_clean = housing_df_clean.astype(float)

# Correlation matrix
corr_matrix = housing_df_clean.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Correlation Matrix of Housing Data")
plt.tight_layout()
plt.show()
```

####  **Linear Regression Models**
Two linear regression models are built:

- **Model 1** includes all numeric variables.
- **Model 2** removes highly collinear variables using Variance Inflation Factor (VIF) analysis.

```python
# Model 1: All predictors
X1 = housing_df_clean.drop('median_house_value', axis=1)
y1 = housing_df_clean['median_house_value']
X1_const = sm.add_constant(X1)
model1 = sm.OLS(y1, X1_const).fit()

# VIF
vif_data = pd.DataFrame()
vif_data["feature"] = X1.columns
vif_data["VIF"] = [variance_inflation_factor(X1.values, i) for i in range(X1.shape[1])]
print(vif_data)
```

####  **Multicollinearity Check (VIF)**
VIF values above 10 indicate multicollinearity, which can distort coefficient estimates. Features like `longitude`, `latitude`, `total_bedrooms`, and `households` show high VIF values, so they are removed in Model 2 to improve interpretability and reduce redundancy.

####  **Residual Plot (Model 2)**
A residuals vs. fitted values plot helps us check for heteroscedasticity — unequal variance in the errors. Ideally, residuals should appear randomly scattered. If there’s a funnel or pattern, it indicates heteroscedasticity, which violates regression assumptions.

```python
# Model 2: Remove high VIF
features_to_keep = vif_data[vif_data["VIF"] < 10]["feature"].tolist()
X2 = X1[features_to_keep]
X2_const = sm.add_constant(X2)
model2 = sm.OLS(y1, X2_const).fit()

# Residuals
residuals = model2.resid
fitted_vals = model2.fittedvalues
sns.residplot(x=fitted_vals, y=residuals, lowess=True)
plt.xlabel("Fitted values")
plt.ylabel("Residuals")
plt.title("Residuals vs Fitted Values (Model 2)")
plt.tight_layout()
plt.show()
```

####  **Heteroscedasticity Test**
We use the **Breusch-Pagan test** to statistically test for heteroscedasticity. A very low p-value confirms its presence, meaning the variance of the residuals is not constant.

####  **Normality Test of Residuals**
The **Shapiro-Wilk test** is used to assess whether residuals are normally distributed. Again, a very low p-value suggests a deviation from normality, although large datasets often lead to statistically significant results even for slight deviations.

####  **Model Comparison using AIC**
The **Akaike Information Criterion (AIC)** is a measure of model quality that balances fit and complexity. Model 1 has a lower AIC, so it is preferred despite the multicollinearity. This indicates it provides a better overall fit to the data.

```python
# Heteroscedasticity test
bp_test = het_breuschpagan(residuals, X2_const)
print("Breusch-Pagan p-value:", bp_test[1])

# Normality test
shapiro_test = shapiro(residuals)
print("Shapiro-Wilk p-value:", shapiro_test[1])

# AIC comparison
print("Model 1 AIC:", model1.aic)
print("Model 2 AIC:", model2.aic)

best_model = model1 if model1.aic < model2.aic else model2
summary_table = pd.concat([best_model.params, best_model.conf_int()], axis=1)
summary_table.columns = ["Coefficient", "CI Lower", "CI Upper"]
print(summary_table)
```

###  Heteroscedasticity Test

- **Test used:** Breusch-Pagan Test  
- **p-value:** `2.26e-174`  
- **Conclusion:** The p-value is extremely low, so we **reject the null hypothesis** of constant variance.  
  This confirms the presence of **heteroscedasticity** in the residuals.

### **Test for Normality of the Residuals**

- **Test used:** Shapiro-Wilk Test  
- **p-value:** `9.45e-72`  
- **Conclusion:** Residuals are **not normally distributed**. This is expected in large datasets and doesn't invalidate the model but may affect inference.


###  **Compare the Two Models Using AIC**

- **Model 1 AIC:** `513118.98`  
- **Model 2 AIC:** `515778.42`  
- **Conclusion:** **Model 1** has a lower AIC and is therefore the better model.

####  **Model Coefficients and Interpretation**
We present the coefficients and their 95% confidence intervals for the best model (Model 1). Interpretation examples:

- `median_income`: For each unit increase in income, the median house value increases by ~$39,259.
- `longitude` and `latitude`: Negative coefficients suggest that housing value decreases as we move west or south.
- `ocean_proximity_INLAND`: Having a house inland reduces its value by ~$39,284 compared to the baseline category.

###  **PART 2: Logistic Regression on Admission Data**

###  Logistic Regression (No Interaction)

We use a logistic regression model to predict whether a student is admitted (`admit = 1`) based on:
- GRE score
- GPA
- Undergraduate institution rank (categorical)

Dummy variables are created for `rank`, with Rank 1 as the reference category. The logistic regression output shows how these variables influence the probability of admission.

The model converged successfully and is statistically significant overall (LLR p-value < 0.001), indicating that at least one predictor is useful.

```python
# Load and clean binary dataset
binary_df = pd.read_csv('binary.csv')
binary_df['rank'] = binary_df['rank'].astype('category')

# Logistic regression without interaction
X_logit = pd.get_dummies(binary_df[['gre', 'gpa', 'rank']], drop_first=True).astype(float)
y_logit = binary_df['admit']
X_logit_const = sm.add_constant(X_logit)

logit_model = sm.Logit(y_logit, X_logit_const).fit()
print(logit_model.summary())

# Odds ratios
print("Odds Ratios:")
print(np.exp(logit_model.params))
```

###  Odds Ratio Interpretation (No Interaction)

To interpret how each variable affects admission, we exponentiate the regression coefficients to obtain **odds ratios**:

- **GRE (coef = 0.0023 → OR ≈ 1.0023):**  
  Each 1-point increase in GRE increases the odds of admission by **~0.23%**, a small but statistically significant effect (*p* = 0.038).

- **GPA (coef = 0.8040 → OR ≈ 2.23):**  
  Each 1-point increase in GPA more than **doubles** the odds of admission, making GPA the most influential predictor (*p* = 0.015).

- **Rank_2 (OR ≈ 0.509):**  
  Students from Rank 2 institutions have **49.1% lower odds** of admission compared to Rank 1 (*p* = 0.033).

- **Rank_3 (OR ≈ 0.262):**  
  Rank 3 applicants have **73.8% lower odds** of admission than Rank 1 (*p* < 0.001).

- **Rank_4 (OR ≈ 0.212):**  
  Rank 4 applicants have **78.8% lower odds** compared to Rank 1 (*p* < 0.001).

###  Interaction Effect: GPA × Rank

We next test for an **interaction effect** between GPA and institutional rank. The question is:

> Does GPA matter more (or less) depending on the prestige of the institution?

We include a new feature `gpa_rank`, calculated by multiplying GPA with the numerical code of the rank (e.g., Rank 1 = 0, Rank 2 = 1, etc.).

```python
# Ensure 'rank' is treated as categorical
binary_df['rank'] = binary_df['rank'].astype('category')

# Create interaction term: GPA * Rank (numeric code)
binary_df['gpa_rank'] = binary_df['gpa'] * binary_df['rank'].cat.codes

# Prepare features: gre, gpa, gpa*rank + dummy variables for rank
X_interact = binary_df[['gre', 'gpa', 'gpa_rank']].join(
    pd.get_dummies(binary_df['rank'], prefix='rank', drop_first=True)
)

# Add constant for intercept
X_interact_const = sm.add_constant(X_interact.astype(float))
y = binary_df['admit']

# Fit logistic regression model
interaction_model = sm.Logit(y, X_interact_const).fit()
print(interaction_model.summary())

# Exponentiate coefficients for odds ratios
print("\nOdds Ratios for Interaction Model:")
print(np.exp(interaction_model.params))
```

###  Odds Ratios for Interaction Model

The new model includes:
- GRE
- GPA
- GPA × Rank (`gpa_rank`)
- Dummy variables for Rank 2, 3, and 4

**Key findings:**

- **GRE (OR ≈ 1.0023)** remains a small but significant predictor.
- **GPA (OR ≈ 2.45)** still shows a strong effect, though now not statistically significant (*p* = 0.104).
- **Interaction term `gpa_rank` (OR ≈ 0.93)** is **not significant** (*p* = 0.831), suggesting that the impact of GPA on admission **does not meaningfully vary** across different institution ranks.
- **All rank dummy variables** become non-significant in the presence of the interaction term.

###  Conclusion

Adding the interaction term between GPA and Rank does **not improve model performance** (Pseudo R² changes marginally from 0.08292 to 0.08301), and none of the interaction-related variables are statistically significant.

The **simpler model without the interaction**:
- Is more interpretable
- Performs similarly
- Shows clearer and statistically significant predictors

Therefore, the original model without interaction is preferred for practical use.
